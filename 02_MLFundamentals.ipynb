```python
import math
import torch
```

```python
D, N = 4, 100
```

```python
x = torch.linspace(-math.pi, math.pi, N)
x
```

```python
y = x.sin()
y
```

```python
def fit_polynomial(D, x, y):

    X = x[:, None] ** torch.arange(0, D + 1)[None]
    # Least square solution
    return torch.linalg.lstsq(X, y).solution
```

```python
alpha = fit_polynomial(D, x, y)
```

```python
X = x[:, None] ** torch.arange(0, D + 1)[None]
```

```python
y_hat = X @ alpha
```
